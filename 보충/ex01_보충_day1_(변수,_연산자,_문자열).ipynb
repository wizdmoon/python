# 🐍 파이썬(Python)이란?

## ✅ 파이썬은 무엇인가요?

파이썬은 **사람과 컴퓨터가 대화할 수 있게 해주는 프로그래밍 언어**입니다.  
쉽고 직관적인 문법 덕분에 **처음 프로그래밍을 배우는 사람도 쉽게 접근할 수 있는 언어**입니다.

> 💬 마치 **요리 레시피처럼** 단계별로 지시를 내리는 방식이에요!  
> 예: “재료를 씻고 → 자르고 → 볶는다”  
> 파이썬: “숫자를 입력받고 → 계산하고 → 결과를 출력한다”

---

## ✅ 왜 파이썬을 배워야 하나요?

- 🔤 **문법이 쉽고 간단해요**  
  → 영어 문장처럼 읽히는 구조

- 🧠 **입문자에게 적합해요**  
  → 복잡한 개념 없이도 시작 가능

- 🛠️ **다양한 분야에 활용돼요**  
  → 웹 개발, 데이터 분석, 인공지능, 자동화, 게임 개발 등

- 🌍 **세계적으로 인기 있는 언어예요**  
  → 개발자 커뮤니티와 자료가 풍부해요

# ➕ 파이썬의 연산자 정리

---

## ✅ 1. 산술 연산자 (Arithmetic Operators)

| 연산자 | 설명 | 예시 | 결과 |
|--------|------|------|------|
| `+` | 덧셈 | `3 + 2` | `5` |
| `-` | 뺄셈 | `5 - 2` | `3` |
| `*` | 곱셈 | `4 * 2` | `8` |
| `/` | 나눗셈 | `5 / 2` | `2.5` |
| `//` | 몫 | `5 // 2` | `2` |
| `%` | 나머지 | `5 % 2` | `1` |
| `**` | 거듭제곱 | `2 ** 3` | `8` |

---

## ✅ 2. 비교 연산자 (Comparison Operators)

| 연산자 | 설명 | 예시 | 결과 |
|--------|------|------|------|
| `==` | 같다 | `3 == 3` | `True` |
| `!=` | 다르다 | `3 != 2` | `True` |
| `>` | 크다 | `5 > 3` | `True` |
| `<` | 작다 | `2 < 1` | `False` |
| `>=` | 크거나 같다 | `3 >= 3` | `True` |
| `<=` | 작거나 같다 | `4 <= 5` | `True` |

---

## ✅ 3. 논리 연산자 (Logical Operators)

| 연산자 | 설명 | 예시 | 결과 |
|--------|------|------|------|
| `and` | 둘 다 참이면 참 | `True and False` | `False` |
| `or` | 둘 중 하나라도 참이면 참 | `True or False` | `True` |
| `not` | 결과를 반대로 | `not True` | `False` |

---

## ✅ 4. 대입 연산자 (Assignment Operators)

| 연산자 | 설명 | 예시 | 결과 |
|--------|------|------|------|
| `=` | 값 대입 | `x = 5` | `x에 5 저장` |
| `+=` | 더해서 대입 | `x += 2` | `x = x + 2` |
| `-=` | 빼서 대입 | `x -= 3` | `x = x - 3` |
| `*=` | 곱해서 대입 | `x *= 2` | `x = x * 2` |
| `/=` | 나눠서 대입 | `x /= 2` | `x = x / 2` |

---

## ✅ 5. 멤버십 연산자 (Membership Operators)

| 연산자 | 설명 | 예시 | 결과 |
|--------|------|------|------|
| `in` | 포함되어 있는가 | `"a" in "apple"` | `True` |
| `not in` | 포함되지 않는가 | `"x" not in "banana"` | `True` |

---

## ✅ 6. 식별 연산자 (Identity Operators)

| 연산자 | 설명 | 예시 | 결과 |
|--------|------|------|------|
| `is` | 같은 객체인가 | `a is b` | `True / False` |
| `is not` | 다른 객체인가 | `a is not b` | `True / False` |

> `is`는 두 값이 같다는 뜻이 아니라, **메모리 주소가 같은 객체인지**를 비교합니다.

---
